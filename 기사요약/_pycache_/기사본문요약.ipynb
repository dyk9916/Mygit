```python
from newspaper import Article

# 파싱할 뉴스 기사 URL 지정
url = 'https://news.samsung.com/kr/ai-%ed%99%88%ec%9c%bc%eb%a1%9c-%ec%9d%bc%ec%83%81%ec%9d%84-%ed%98%81%ec%8b%a0%ed%95%98%eb%8b%a4-%ec%82%bc%ec%84%b1%ec%a0%84%ec%9e%90-2025%eb%85%84%ed%98%95-%eb%b9%84%ec%8a%a4'

# 언어를 한국어로 설정하고 URL을 전달해 Article 클래스의 객체 생성
article = Article(url, language='ko')

# 지정된 웹 페이지를 다운로드
article.download()

# 다운로드한 웹 페이지를 분석하고 필요한 정보를 추출
article.parse()
```

```python
# 기사 제목 출력
print('기사 제목 :',article.title)
print('')

# 기사 내용 출력 
print('기사 내용 :') 
print(article.text)
```

```python
# 기사 제목과 내용을 텍스트 파일로 저장
with open('article.txt', 'w', encoding='utf-8') as f:
  f.write('기사 제목 : ' + article.title + '\n\n')
  f.write('기사 내용 :\n' + article.text)
```

```python
import openai

# OpenAI API 키 입력
client = openai.OpenAI(api_key="")  # 본인 키로 교체

# 기사 본문 불러오기
with open('article.txt', 'r', encoding='utf-8') as f:
    article_text = f.read()

# 프롬프트 생성
prompt = f"다음 기사 내용을 한국어로 5문장 이내로 핵심만, 너무 딱딱하지 않고 부드럽고 친절한 어투로 요약해줘:\n\n{article_text}"

# GPT API 호출 (openai 1.0 버전)
response = client.chat.completions.create(
    model="gpt-4-turbo",
    messages=[
        {"role": "system", "content": "당신은 친절하고 부드러운 기사 요약가입니다."},
        {"role": "user", "content": prompt}
    ],
    max_tokens=500,
    temperature=0.5,
)

# 요약 결과 출력
summary = response.choices[0].message.content
print("요약 결과:\n", summary)
```

```python
import os

output_dir = 'output'
os.makedirs(output_dir, exist_ok=True)

with open(os.path.join(output_dir, '삼성전자, 2025년형 비스포크 AI 라인업 공개.txt'), 'w', encoding='utf-8') as f:
    f.write('기사 요약:\n' + summary)
```
