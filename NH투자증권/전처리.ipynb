```python
# Remove leading and trailing spaces from '티커종목코드' column
df['티커종목코드'] = df['티커종목코드'].str.strip()

# Verify the changes
df['티커종목코드'].unique()
```

```python
# Define a function to create QA pairs using the provided template
def generate_qa(df):
    qa_pairs = []
    for index, row in df.iterrows():
        ticker = row['티커종목코드']
        classification = row['고객구성중분류코드']
        avg_account_ratio = round(row['평균고객계좌수비율'], 2)
        avg_investment_ratio = round(row['평균고객투자비율'], 2)
        
        # Create QA pair for customer account ratio
        qa_account = {
            "question": f"What is the average customer account ratio for {classification} for ticker {ticker}?",
            "answer": f"The average customer account ratio is {avg_account_ratio}%."
        }
        qa_pairs.append(qa_account)
        
        # Create QA pair for customer investment ratio
        qa_investment = {
            "question": f"What is the average investment ratio for {classification} for ticker {ticker}?",
            "answer": f"The average investment ratio is {avg_investment_ratio}%."
        }
        qa_pairs.append(qa_investment)
    
    return qa_pairs

# Generate QA data
qa_data = generate_qa(df)

# Display first few QA pairs as an example
qa_data[:5]

# Convert the combined QA data into a dataframe with two columns: 'question' and 'answer'
qa_df = pd.DataFrame(qa_data)
```

```python
# Function to generate QA data for the question "What age group invested the most in ~ stocks?"
def generate_age_group_investment_qa(df):
    qa_pairs = []
    
    # Filter the dataset for age groups only
    age_group_df = df[df['고객구성중분류코드'].str.contains('세')]

    # Group by ticker and find the age group with the maximum investment ratio for each ticker
    max_investment_by_age_group = age_group_df.groupby('티커종목코드').apply(
        lambda x: x.loc[x['평균고객투자비율'].idxmax()]
    ).reset_index(drop=True)

    # Create QA pairs
    for index, row in max_investment_by_age_group.iterrows():
        ticker = row['티커종목코드'].strip()
        age_group = row['고객구성중분류코드']
        qa_pair = {
            "question": f"What age group invested the most in {ticker} stocks?",
            "answer": f"The {age_group} age group invested the most."
        }
        qa_pairs.append(qa_pair)
    
    return qa_pairs

# Generate the age group investment QA data
age_group_investment_qa = generate_age_group_investment_qa(df)

# Display first few QA pairs as an example
age_group_investment_qa[:5]

# Convert the combined QA data into a dataframe with two columns: 'question' and 'answer'
qa_df_age = pd.DataFrame(age_group_investment_qa)
```

```python
# Function to generate QA data for the question "What age group invested the most in ~ stocks?"
def generate_perfomance_group_investment_qa(df):
    qa_pairs = []
    
    # Filter the dataset for age groups only
    age_group_df = df[df['고객구성중분류코드'].str.contains('투자')]

    # Group by ticker and find the age group with the maximum investment ratio for each ticker
    max_investment_by_age_group = age_group_df.groupby('티커종목코드').apply(
        lambda x: x.loc[x['평균고객투자비율'].idxmax()]
    ).reset_index(drop=True)

    # Create QA pairs
    for index, row in max_investment_by_age_group.iterrows():
        ticker = row['티커종목코드'].strip()
        age_group = row['고객구성중분류코드']
        qa_pair = {
            "question": f"Who invested the most in {ticker} stocks?",
            "answer": f"{age_group} invested the most."
        }
        qa_pairs.append(qa_pair)
    
    return qa_pairs

# Generate the age group investment QA data
performance_group_investment_qa = generate_perfomance_group_investment_qa(df)

# Convert the combined QA data into a dataframe with two columns: 'question' and 'answer'
qa_df_performance = pd.DataFrame(performance_group_investment_qa)
```

```python
# Function to generate QA data for the question "Stocks with what size of assets were the most invested?"
def generate_asset_size_investment_qa(df):
    qa_pairs = []
    
    # Filter the dataset for asset size groups only
    asset_size_df = df[df['고객구성중분류코드'].str.contains('억원|만원')]

    # Group by ticker and find the asset size with the maximum investment ratio for each ticker
    max_investment_by_asset_size = asset_size_df.groupby('티커종목코드').apply(
        lambda x: x.loc[x['평균고객투자비율'].idxmax()]
    ).reset_index(drop=True)

    # Create QA pairs
    for index, row in max_investment_by_asset_size.iterrows():
        ticker = row['티커종목코드'].strip()
        asset_size = row['고객구성중분류코드']
        qa_pair = {
            "question": f"{ticker} stocks with what size of assets were the most invested?",
            "answer": f"Those with {asset_size} amount of assets invested the most."
        }
        qa_pairs.append(qa_pair)
    
    return qa_pairs

# Generate the asset size investment QA data
asset_size_investment_qa = generate_asset_size_investment_qa(df)

# Display first few QA pairs as an example
asset_size_investment_qa[:20]

# Convert the combined QA data into a dataframe with two columns: 'question' and 'answer'
qa_df_asset_size = pd.DataFrame(asset_size_investment_qa)
```

```python
# Function to generate combined QA pairs
def generate_combined_qa(df):
    qa_pairs = []
    
    # Get unique tickers
    tickers = df['티커종목코드'].unique()
    
    for ticker in tickers:
        # Filter data for the current ticker
        ticker_df = df[df['티커종목코드'] == ticker]
        
        # Filter for age groups (assuming age groups contain '세')
        age_group_df = ticker_df[ticker_df['고객구성중분류코드'].str.contains('세')]
        if not age_group_df.empty:
            # Find the age group with the highest investment ratio
            top_age_group = age_group_df.loc[age_group_df['평균고객투자비율'].idxmax()]['고객구성중분류코드']
        else:
            top_age_group = "정보 없음"
        
        # Filter for asset sizes (assuming asset sizes contain '만원' or '억원')
        asset_size_df = ticker_df[ticker_df['고객구성중분류코드'].str.contains('만원|억원')]
        if not asset_size_df.empty:
            # Find the asset size with the highest investment ratio
            top_asset_size = asset_size_df.loc[asset_size_df['평균고객투자비율'].idxmax()]['고객구성중분류코드']
        else:
            top_asset_size = "정보 없음"
        
        # Create the QA pair
        qa_pair = {
            "question": f"Please tell me about the customer group that purchased {ticker} stocks.",
            "answer": f"{ticker} stocks are most held by customers of the {top_age_group} age group, and customers with {top_asset_size} asset size are the largest."
        }
        qa_pairs.append(qa_pair)
    
    return qa_pairs

# Generate the combined QA data
combined_qa_data = generate_combined_qa(df)

# Display first few QA pairs as an example
combined_qa_data[:5]

# Convert the combined QA data into a dataframe with two columns: 'question' and 'answer'
qa_df_combined = pd.DataFrame(combined_qa_data)
```

```python
# Divide the data into two categories based on '주식/ETF구분코드'
stock_data = data[data['주식/ETF구분코드'] == '주식']
etf_data = data[data['주식/ETF구분코드'] == 'ETF']
```

```python
# Create a function that formats the data into QA format for each company
def create_qa_format(row):
    question = f"Tell me about {row['외화증권영문명']}."
    answer = (
        f"{row['외화증권영문명']}’s ticker stock code is {row['티커종목코드']}, "
        f"the industry is {row['산업명']}, {row['영문사업개요내용']}, "
        f"and the CEO is {row['CEO명']}, market capitalization is {row['시가총액']}."
    )
    return {"question": question, "answer": answer}

# Apply the function to the stock data and create a new dataframe
qa_stock_data = stock_data.apply(create_qa_format, axis=1)

# Convert the result to a list of dictionaries for easier handling
qa_stock_data_list = qa_stock_data.tolist()

# Display the first few Q&A pairs to the user
qa_stock_data_list[:5]

# Convert the combined QA data into a dataframe with two columns: 'question' and 'answer'
qa_stock_data = pd.DataFrame(qa_stock_data_list)
```
