### 석식, 중식 기준으로 데이터셋 분할 

```python
# 중식과 석식 관련 컬럼을 각각 분리
lunch_cols = ['일자', '요일', '중식계', 'month', 'corona', 'day_gap', '중식_국', '중식_메인1', '중식_메인2',
              '출근', '휴가비율', '출장비율', '야근비율', '재택비율']
dinner_cols = ['일자', '요일', '석식계', 'month', 'corona', 'day_gap', '석식_국', '석식_메인1', '석식_메인2',
               '출근', '휴가비율', '출장비율', '야근비율', '재택비율']

lunch_cols_for_test = ['일자', '요일', 'month', 'corona', 'day_gap', '중식_국', '중식_메인1', '중식_메인2',
              '출근', '휴가비율', '출장비율', '야근비율', '재택비율']
dinner_cols_for_test = ['일자', '요일', 'month', 'corona', 'day_gap', '석식_국', '석식_메인1', '석식_메인2',
               '출근', '휴가비율', '출장비율', '야근비율', '재택비율']

# lunch와 dinner 데이터셋 나누기
lunch_train = train[lunch_cols].copy()
dinner_train = train[dinner_cols].copy()

lunch_test = test[lunch_cols_for_test].copy()
dinner_test = test[dinner_cols_for_test].copy()
```

```python
# 4. 중식과 석식의 메뉴 텍스트 결합
menu_columns = {
    '중식': ['중식_국', '중식_메인1', '중식_메인2'],
    '석식': ['석식_국', '석식_메인1', '석식_메인2']
}

# 결측치 처리 및 문장 생성
lunch_train['중식'] = lunch_train[menu_columns['중식']].fillna('').agg(' '.join, axis=1)
dinner_train['석식'] = dinner_train[menu_columns['석식']].fillna('').agg(' '.join, axis=1)
```

```python
lunch_train.head()
```

```python
# 결측치 처리 및 문장 생성
lunch_test['중식'] = lunch_test[menu_columns['중식']].fillna('').agg(' '.join, axis=1)
dinner_test['석식'] = dinner_test[menu_columns['석식']].fillna('').agg(' '.join, axis=1)
```

```python
lunch_test.head()
```

```python
from gensim.models import Word2Vec

# 1. train/test의 모든 메뉴(중식/석식)를 하나의 리스트로 모으기
def get_menu_lists(df, menu_cols):
    # 각 행의 메뉴(국, 메인1, 메인2)를 리스트로 합침
    return df[menu_cols].fillna('').values.tolist()

menu_cols_lunch = ['중식_국', '중식_메인1', '중식_메인2']
menu_cols_dinner = ['석식_국', '석식_메인1', '석식_메인2']

# 모든 메뉴를 하나의 리스트로
all_menus = []
for df in [train, test]:
    for cols in [menu_cols_lunch, menu_cols_dinner]:
        for row in df[cols].fillna('').values:
            # 빈 값 제거
            tokens = [str(x) for x in row if x]
            if tokens:
                all_menus.append(tokens)

# 2. Word2Vec 학습
w2v_model = Word2Vec(sentences=all_menus, vector_size=1, window=3, min_count=1, workers=4, sg=1, seed=42)

# 3. 각 샘플별로 메뉴 임베딩의 평균을 구하는 함수
def get_mean_embedding(row, model, cols):
    tokens = [str(row[col]) for col in cols if pd.notnull(row[col]) and row[col] in model.wv]
    if not tokens:
        return np.zeros(model.vector_size)
    vectors = [model.wv[token] for token in tokens]
    return np.mean(vectors, axis=0)

# 4. lunch_train, dinner_train에 feature로 추가
for i in range(w2v_model.vector_size):
    lunch_train[f'w2v_lunch_{i}'] = lunch_train.apply(lambda row: get_mean_embedding(row, w2v_model, menu_cols_lunch)[i], axis=1)
    dinner_train[f'w2v_dinner_{i}'] = dinner_train.apply(lambda row: get_mean_embedding(row, w2v_model, menu_cols_dinner)[i], axis=1)
    
# 5. lunch_test, dinner_test에 feature로 추가
for i in range(w2v_model.vector_size):
    lunch_test[f'w2v_lunch_{i}'] = lunch_test.apply(lambda row: get_mean_embedding(row, w2v_model, menu_cols_lunch)[i], axis=1)
    dinner_test[f'w2v_dinner_{i}'] = dinner_test.apply(lambda row: get_mean_embedding(row, w2v_model, menu_cols_dinner)[i], axis=1)
```
