```python
df_copy = df.copy()
df_copy
```

```python
changed_columns=['FTHG', 'FTAG',
       'HTHG', 'HTAG', 'HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC', 'AC',
       'HY', 'AY', 'HR', 'AR', 'HxG', 'AxG', 'HxA', 'AxA', 'HxPTS', 'AxPTS',
       'HPPDA', 'APPDA']
```

```python
HomeTeam_Continuous=['FTHG', 'HTHG', 'HS', 'HST', 'HF', 'HC', 'HY', 'HR', 'HxG', 'HxA', 'HxPTS', 'HPPDA']
AwayTeam_Continuous=['FTAG', 'HTAG', 'AS', 'AST', 'AF', 'AC', 'AY', 'AR', 'AxG', 'AxA', 'AxPTS', 'APPDA']
```

```python
# 홈팀인경우변환
for i in range(len(df)):

    temp=0
    temp_df = pd.DataFrame(columns=HomeTeam_Continuous)
    for j in range(i+1,len(df)):

        if df.iloc[i]['HomeTeam'] == df.iloc[j]["HomeTeam"]:

            temp_df=pd.concat([temp_df, df.loc[j:j,HomeTeam_Continuous]], axis=0)
            temp+=1
        if df.iloc[i]['HomeTeam'] == df.iloc[j]["AwayTeam"]:

            temp_away=df.loc[j:j,AwayTeam_Continuous]
            temp_away.columns=HomeTeam_Continuous
            temp_df=pd.concat([temp_df, temp_away], axis=0)
            temp+=1
        if temp == 5:
            break
    temp_mean=temp_df.mean(axis=0)
    temp_mean=pd.DataFrame(temp_mean)
    temp_mean=temp_mean.transpose()
    df_copy.loc[i,HomeTeam_Continuous]= temp_mean.loc[0,HomeTeam_Continuous]
# 어웨이팀인경우변환
for i in range(len(df)):

    temp=0
    temp_df = pd.DataFrame(columns=AwayTeam_Continuous)
    for j in range(i+1,len(df)):

        if df.iloc[i]['AwayTeam'] == df.iloc[j]["AwayTeam"]:

            temp_df=pd.concat([temp_df, df.loc[j:j,AwayTeam_Continuous]], axis=0)
            temp+=1
        if df.iloc[i]['AwayTeam'] == df.iloc[j]["HomeTeam"]:

            temp_home=df.loc[j:j,HomeTeam_Continuous]
            temp_home.columns=AwayTeam_Continuous
            temp_df=pd.concat([temp_df, temp_home], axis=0)
            temp+=1
        if temp == 5:
            break
    temp_mean=temp_df.mean(axis=0)
    temp_mean=pd.DataFrame(temp_mean)
    temp_mean=temp_mean.transpose()
    df_copy.loc[i,AwayTeam_Continuous]= temp_mean.loc[0,AwayTeam_Continuous]



```
