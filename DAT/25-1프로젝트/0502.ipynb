```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('D:/김동영/11_Github/mygit-1/DAT/25-1프로젝트/소셜핸들.xlsx')
df1 = pd.read_csv('D:/김동영/11_Github/mygit-1/DAT/25-1프로젝트/exolyt_result4.csv')
df2 = pd.read_csv('D:/김동영/11_Github/mygit-1/DAT/25-1프로젝트/exolyt_result5.csv')
```

```python
df
df['Username'] = df['Username'].str.replace('@', '', regex=False)
```

```python
df_merged = df.merge(df1, left_on='Username', right_on='username', how='left') \
        .merge(df2, left_on='Username', right_on='username', how='left', suffixes=('_likes', '_videos'))

df_merged = df_merged.drop(columns=['username_likes', 'username_videos'])
```

```python
df_merged.to_csv('D:/김동영/11_Github/mygit-1/DAT/25-1프로젝트/최종데이터/좋아요&영상수.csv', index=False)
```

```python
data = pd.read_csv('D:/김동영/11_Github/mygit-1/DAT/25-1프로젝트/최종데이터/final_data.csv')
```

```python
merged_data = data.merge(df_merged, left_on='Username', right_on='Username', how='left')
merged_data
```

```python
merged_data = merged_data.dropna()
merged_data
```
