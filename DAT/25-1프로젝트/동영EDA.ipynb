```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# 1. 크롬 드라이버 설정
options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_argument('--disable-notifications')
options.add_argument('--disable-popup-blocking')
# options.add_argument('--headless')  # ← 일단 꺼두고 테스트

# 드라이버 실행
driver = webdriver.Chrome(options=options)

try:
    # 2. 페이지 접속
    url = "https://exolyt.com/user/tiktok/xevelynluu"
    driver.get(url)

    # 3. 페이지 로딩 대기
    WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.XPATH, "//h2[contains(text(), 'Overview')]"))
    )
    time.sleep(2)  # 렌더링 보조 시간

    # 4. 정확한 위치의 수치 크롤링 (5.7M 해당)
    target_elem = WebDriverWait(driver, 20).until(
        EC.presence_of_element_located(
            (By.XPATH, "(//div[contains(@class, 'text-lg') and contains(@class, 'font-bold')])[1]")
        )
    )

    print("📊 추출된 수치:", target_elem.text)

except Exception as e:
    print("❌ 크롤링 실패:", e)

finally:
    driver.quit()
```

```python
try:
    # 2. Exolyt 사용자 페이지 열기
    driver.get("https://exolyt.com/user/tiktok/xevelynluu")

    # 3. 페이지 로딩 대기
    WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.XPATH, "//h2[contains(text(), 'Overview')]"))
    )
    time.sleep(2)

    # 4. Avg. Views 크롤링
    avg_views_elem = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located(
            (By.XPATH, "//div[contains(text(), 'Avg. Views')]/following-sibling::div")
        )
    )
    avg_views = avg_views_elem.text

    # 5. Avg. Comments 크롤링
    avg_comments_elem = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located(
            (By.XPATH, "//div[contains(text(), 'Avg. Comments')]/following-sibling::div")
        )
    )
    avg_comments = avg_comments_elem.text

    # 6. 결과 출력
    print("📊 Avg. Views:", avg_views)
    print("💬 Avg. Comments:", avg_comments)

except Exception as e:
    print("❌ 크롤링 실패:", e)

finally:
    driver.quit()
```
