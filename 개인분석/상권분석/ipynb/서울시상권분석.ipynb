```python
import xml.etree.ElementTree as ET
```

```python
# 응답 내용 출력
print("응답 상태 코드:", res.status_code)
print("응답 헤더:", res.headers)
print("응답 내용:", res.text)
```

```python
# XML 응답 처리
if res.headers.get("Content-Type", "").startswith("application/xml"):
    root = ET.fromstring(res.text)
    
    # 특정 태그만 추출 (예: 'STDR_YYQU_CD', 'ADSTRD_CD_NM', 'THSMON_SELNG_AMT')
    for row in root.findall(".//row"):
        year_quarter = row.find("STDR_YYQU_CD").text  # 기준 연도 분기
        district_name = row.find("ADSTRD_CD_NM").text  # 행정동 이름
        monthly_sales = row.find("THSMON_SELNG_AMT").text  # 월 매출액
        
        print(f"기준 연도 분기: {year_quarter}, 행정동 이름: {district_name}, 월 매출액: {monthly_sales}")
else:
    print("응답이 XML 형식이 아닙니다.")
```

```python
# API URL
url = "http://openapi.seoul.go.kr:8088/556648745364796b38345043555455/xml/VwsmAdstrdSelngW/1/5/"
res = requests.get(url)

# XML 응답 처리
if res.headers.get("Content-Type") == "application/xml":
    root = ET.fromstring(res.text)
    
    # 특정 태그만 추출 (예: 'STDR_YYQU_CD', 'ADSTRD_CD_NM', 'THSMON_SELNG_AMT')
    for row in root.findall(".//row"):
        year_quarter = row.find("STDR_YYQU_CD").text  # 기준 연도 분기
        district_name = row.find("ADSTRD_CD_NM").text  # 행정동 이름
        monthly_sales = row.find("THSMON_SELNG_AMT").text  # 월 매출액
        
        print(f"기준 연도 분기: {year_quarter}, 행정동 이름: {district_name}, 월 매출액: {monthly_sales}")
else:
    print("응답이 XML 형식이 아닙니다.")
    print(res.text)
```

```python
# API URL 기본 설정
base_url = "http://openapi.seoul.go.kr:8088/556648745364796b38345043555455/xml/VwsmAdstrdSelngW"
start_index = 1
end_index = 1000  # 최대 요청 가능 건수
all_data = []  # 모든 데이터를 저장할 리스트

while True:
    # URL에 시작 위치와 종료 위치를 추가
    url = f"{base_url}/{start_index}/{end_index}/"
    res = requests.get(url)
    
    # XML 응답 처리
    if res.headers.get("Content-Type", "").startswith("application/xml"):
        root = ET.fromstring(res.text)
        
        # 에러 코드 확인
        error_code = root.find(".//CODE")
        if error_code is not None and error_code.text != "INFO-000":
            print("더 이상 데이터를 가져올 수 없습니다.")
            break
        
        # <row> 태그 데이터 추출
        rows = root.findall(".//row")
        if not rows:  # 더 이상 데이터가 없으면 종료
            print("모든 데이터를 가져왔습니다.")
            break
        
        # 데이터를 리스트에 추가
        for row in rows:
            year_quarter = row.find("STDR_YYQU_CD").text  # 기준 연도 분기
            district_name = row.find("ADSTRD_CD_NM").text  # 행정동 이름
            service_name = row.find("SVC_INDUTY_CD_NM").text # 서비스업종명
            monthly_sales = row.find("THSMON_SELNG_AMT").text  # 월 매출액
            weekday_sales = row.find("MDWK_SELNG_AMT").text # 주중 매출액
            weekend_sales = row.find("WKEND_SELNG_AMT").text  # 주말 매출액
            monday_sales = row.find("MON_SELNG_AMT").text # 월요일 매출액
            tuesday_sales = row.find("TUES_SELNG_AMT").text # 화요일 매출액
            wednesday_sales = row.find("WED_SELNG_AMT").text # 수요일 매출액
            thursday_sales = row.find("THUR_SELNG_AMT").text # 목요일 매출액
            friday_sales = row.find("FRI_SELNG_AMT").text # 금요일 매출액
            saturday_sales = row.find("SAT_SELNG_AMT").text # 토요일 매출액
            sunday_sales = row.find("SUN_SELNG_AMT").text # 일요일 매출액
            male_sales = row.find("ML_SELNG_AMT").text # 남성 매출액
            female_sales = row.find("FML_SELNG_AMT").text # 여성 매출액
            age_10_sales = row.find("AGRDE_10_SELNG_AMT").text # 10대 매출액
            age_20_sales = row.find("AGRDE_20_SELNG_AMT").text # 20대 매출액
            age_30_sales = row.find("AGRDE_30_SELNG_AMT").text # 30대 매출액
            age_40_sales = row.find("AGRDE_40_SELNG_AMT").text # 40대 매출액
            age_50_sales = row.find("AGRDE_50_SELNG_AMT").text # 50대 매출액
            age_over_60_sales = row.find("AGRDE_60_ABOVE_SELNG_AMT").text # 60대 이상 매출액
            
            
            
            all_data.append({
                "기준 연도 분기": year_quarter,
                "행정동 이름": district_name,
                "서비스업종명": service_name,
                "월 매출액": monthly_sales,
                "주중 매출액": weekday_sales,
                "주말 매출액": weekend_sales,
                "월요일 매출": monday_sales,
                "화요일 매출": tuesday_sales,
                "수요일 매출": wednesday_sales,
                "목요일 매출": thursday_sales,
                "금요일 매출": friday_sales,
                "토요일 매출": saturday_sales,
                "일요일 매출": sunday_sales,
                "남성 매출액": male_sales,
                "여성 매출": female_sales,
                "10대 매출": age_10_sales,
                "20대 매출": age_20_sales,
                "30대 매출": age_30_sales,
                "40대 매출": age_40_sales,
                "50대 매출": age_50_sales,
                "60대 이상 매출": age_over_60_sales
            })
        
        # 다음 요청 범위 설정
        start_index += 1000
        end_index += 1000
    else:
        print("응답이 XML 형식이 아닙니다.")
        print(res.text)
        break

# 총 데이터 개수 출력
print(f"총 데이터 개수: {len(all_data)}")

# 데이터 출력 (필요한 경우)
for data in all_data[:10]:  # 처음 10개만 출력
    print(data)
```
